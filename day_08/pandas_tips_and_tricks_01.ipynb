# how to rename a column

```python
pd.DataFrame(np.random.rand(5, 10), columns=['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J'], index=['A', 'B', 'C', 'D', 'E'])
df2 = pd.DataFrame(np.random.rand(5, 10), columns=['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J'], index=['A', 'B', 'C', 'D', 'E'])
print(df2)
```

```python
df2.rename(columns={'A':'a', 'B':'b', 'C':'c', 'D':'d', 'E':'e', 'F':'f', 'G':'g', 'H':'h', 'I':'i', 'J':'j'}, inplace=True)
print(df2)
#use this method when you want to change name of a specific column.
```

```python
#another way is to simply give new names to the columns as follows:
df2.columns=['A_col', 'B_col', 'C_col', 'D_col', 'E_col', 'F_col', 'G_col', 'H_col', 'I_col', 'J_col']
df2
```

```python
#if we want to chnage the special character to something else, we can do it by using the following
df2.columns=df2.columns.str.replace('_', '*')
df2
```

```python
#adding prefix or suffix
df2=df2.add_prefix('Rocky')
df2=df2.add_suffix('_Rocky')

#Running this will add prefix or sufix
```

# Filtring a dataset

```python
import seaborn as sns
import pandas as pd

df=sns.load_dataset('titanic')
```

```python
df.head()
```

```python
#if we want to for example see if there are any other genders that are listed in the data, we can do so by

df.sex.unique()
```

```python
df[(df.sex=='female')]
```

```python
df.pclass.unique()
```

```python
df.embark_town.unique()
```

```python
df[(df.embark_town=='Queenstown')]
```

```python
#if you want to see two specific categories together do this. Here "&" is used to combine two categories. For example, if you want to see the categories "A" and "B" together, you can use "&A&B" in the category field. 
# If you want to see the categories "A" and "B" and "C" together, you can use "&A&B&C" in the category field together.
df[(df.embark_town == 'Queenstown') & (df.sex == 'female')]
```

```python
#if you want to apply function that one of the two categories is present i.e. either category1 or category2 is present then you can use "|" (OR) operator.
df[(df.embark_town=='Southampton') | (df.embark_town=='Cherbourg')]
```

```python
#You can also combine '&' and '|' (Or) operators to create more complex conditions.

df[((df.embark_town=='Queenstown') | (df.embark_town=='Southampton')) & (df.sex=='male')]

#adding an extra bracket with embark_town is to make sure the conditions are evaluated in the correct order. Without it, the condition would be evaluated as (df.embark_town=='Queenstown' | df.sex=='male') & (df .embark_town=='Southampton')
#which may lead to addition of females from Southampton to the result.
```

```python
#another method which can be used on pandas for specific category filtering is using 'isin' function which is used to filter the data based on the condition that the value should be present in the list of values. Here is an example of how to use it:
df[df.embark_town.isin(['Queenstown', 'Southampton'])].head()
```

```python
df[df.age<35]
```

```python
df[(df['age'] >= 15) & (df['age'] <= 30)]
```

# Reshaping a dataframe

```python
distance = pd.DataFrame([['12345', 100, 200, 300], ['34567', 400, 500, 600], ['67890', 700, 800, 900]], 
                        columns=['zip', 'factory', 'warehouse', 'retail'])
distance.head()
```

```python
#this converts a wider format data into long format dataset by "melting" multiple columns into key-value pairs.
#id_vars is identifier vairable i.e, the variable that we want to keep as it is.
#var_name is the name of the variable that we want to melt.
#value_name is the name of the variable that we want to create after melting.

distance.melt(id_vars='zip', var_name='location_type', value_name='distance')
distance1= distance.melt(id_vars='zip', var_name='location_type', value_name='distance')
distance1.head(10)
```

```python
sns.barplot(x='zip', y='distance', hue='location_type', data=distance1)
```
