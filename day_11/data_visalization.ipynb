```python
#import libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#import dataset
titanic= sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
titanic.info()
```

```python
titanic.isnull()
```

```python
titanic.isnull().sum()
```

```python
titanic.survived.value_counts().plot(kind='bar')
plt.title('Survived vs Not Survived')
```

```python
sns.countplot(x='survived', hue= 'sex', data=titanic)
```

```python
#scatter plot
plt.scatter(data=titanic, x='age', y='fare', alpha=0.5, c='pclass', cmap='viridis')
plt.title('Age vs Fare')

#here we are using scatter plot to see the relationship between age and fare
#alpha is used to set the transparency of the points
#c is used to set the color of the points based on pclass
#cmap is used to set the color map
#here we are using viridis color map which is a perceptually uniform color map
#this means that the colors are evenly spaced and can be easily distinguished by the human eye
#we can see that the points are scattered and there is no clear relationship between age and fare

```

```python
#scatter plot
sns.scatterplot(data=titanic, x='age', y='fare', hue='pclass')

#we can use hue to set the color of the points based on pclass
#her we have used sns scatterplot to plot the data where as in the previous example we used plt.scatter
#sns scatterplot is a high level interface for drawing attractive and informative statistical graphics
#we can see that the points are scattered and there is no clear relationship between age and fare
```

```python
#bar plot
sns.barplot(data=titanic, x='sex', y='fare', hue='pclass')
```

```python
#pair plot
sns.pairplot(titanic)

#in this plot we can see the relationship between all the features in the dataset in a single plot
```

```python
#histplot
sns.histplot(titanic)
```

```python
#or another way to do it by
titanic.hist(figsize=(10,10), bins=10, color='red', alpha=0.5, grid=False)
#bins is used to set the number of bins in the histogram. bins is the number of intervals in which the data is divided
#color is used to set the color of the histogram
#grid is used to set the grid lines in the histogram
#we can see that the histogram is not normally distributed and is skewed to the right
```

```python
#box plot
sns.boxplot(data=titanic, x='sex', y='fare', hue='pclass')
#in this plot we can see the relationship between sex and fare in different classes
#this plot helps to visualize the distribution of fare across different classes for each gender
#we can also add a title to the plot for better understanding
plt.title('Fare Distribution')
```

```python
sns.boxenplot(data=titanic, x='sex', y='fare', hue='pclass')
#in this plot we can see the relationship between sex and fare for each class of pclass. The boxenplot is a combination of boxplot and violin plot
```

```python
#for changing the title of the plot we can use plt.title() function
#we can also use sns.set_title() function to set the title of the plot
plt.title('Plot of x vs y')
#we can also use sns.set() function to set the title of the plots
sns.set(title='Plot of x vs y')
```

```python
g= sns.FacetGrid(titanic, hue='survived', col='pclass', margin_titles=True,
                 palette={1:'seagreen', 0:'gray'})
g= g.map(plt.scatter, 'age', 'fare', edgecolor='w').add_legend()
#in this plot we can see the relationship between age and fare for each class of pclass. The FacetGrid is a combination of scatterplot and boxplot . 
# The hue parameter is used to color the points based on the survived column. The col parameter is used to create separate plots for each class of pclass. The margin_titles parameter is used to set the title of the plot. The palette parameter is used to set the color of the points based on the survived column. The edgecolor parameter is used to set the color of the edges of the points. The add_legend() function is used to add a legend to the plot.
#we can also use sns.set() function to set the title of the plots . For example, sns.set(title='Titanic Survivors') will set the title of the plots to 'Titanic Survivors'
#we can also use sns.set_style() function to set the style of the plots. For example, sns.set_style('darkgrid') will set the style of the plots to darkgrid. The available styles are 'darkgrid', 'whitegrid', 'dark', 'white', and 'ticks'. The default style is 'darkgrid'.
```

```python
g= sns.FacetGrid(titanic, hue='survived', col='sex', margin_titles=True,
                 palette={1:'seagreen', 0:'gray'})
g= g.map(plt.scatter, 'age', 'fare', edgecolor='w').add_legend()
```

```python
#strip plot
sns.stripplot(data=titanic, x='sex', y='age', hue='survived')
#in this plot we can see the relationship between sex and age and survival status of the passengers. 
# The plot shows that the female passengers were younger than the male passengers and the survival rate of the female passengers was higher than the male passengers.
plt.title('Age Distribution of Passengers')
```

```python
# Convert categorical variables to numeric
titanic_encoded = pd.get_dummies(titanic, drop_first=True)

# Recompute the correlation matrix
corr = titanic_encoded.corr()
plt.figure(figsize=(10,10))
sns.heatmap(corr)
plt.title('Correlation Matrix of Titanic Dataset')
```

```python
#heatmap
# Convert categorical variables to numeric
titanic_encoded = pd.get_dummies(titanic, drop_first=True)

# Recompute the correlation matrix
corr = titanic_encoded.corr()
plt.figure(figsize=(10,10))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix of Titanic Dataset')
#in this plot we can see the correlation between all the features in the dataset. 
# The annot parameter is used to show the correlation values in the heatmap. 
# The cmap parameter is used to set the color map of the heatmap. 
# The fmt parameter is used to set the format of the correlation values in the heatmap i.e., .2f means 2 decimal places.
# The title parameter is used to set the title of the heatmap.
```

```python
#violin plot

sns.violinplot(data=titanic, x='sex', y='age', hue='pclass', split=True)
#split=True is used to split the violin plot into two halves based on the hue parameter.
#in this plot we can see the relationship between age and sex, differentiated by passenger class.
plt.title('Violin Plot of Age by Sex and Passenger Class')
plt.xlabel('Sex')
plt.ylabel('Age')
plt.show()
```
