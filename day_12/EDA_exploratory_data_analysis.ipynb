# Dealing with missing values in EDA analysis

#1. drop the column\
df.drop(columns=['deck'],inplace=True)\
#2. drop the rows\
df.dropna(subset=['age'],inplace=True)\
#3. impute the values with mean or median or mode\
df['age'].fillna(df['age'].mean(),inplace=True)\
#4. impute the values with mean or median or mode based on the categorical variable\
df['embarked'].fillna(df['embarked'].mode()[0],inplace=True)\
#5. impute the values with mean or median or mode based on the categorical variable and group by the categorical variable\
df['age'].fillna(df.groupby('pclass')['age'].transform('mean'),inplace=True)\
#6. impute the values with mean or median or mode based on the categorical variable and group by the categorical variable and drop the rows with missing values\
df.dropna(subset=['age'],inplace=True)\
#7. impute the values with mean or median or mode based on the categorical variable and group by the categorical variable and drop the rows with missing values and drop the column with missing values\
df.drop(columns=['deck'],inplace=True)\
#8 imputing NaN values with 0\
df.fillna(0,inplace=True)\
#9 imputing NaN values with 0 based on the categorical variable and group by the categorical variable\
