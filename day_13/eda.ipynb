# explaining boxplot

a box plot consists of the median value of the data, and the first and third quartiles (Q1 and Q3). Any data points that fall outside of 1.5 times the interquartile range (IQR) i.e., 1.5 * (Q3 - Q1) are considered outliers. The box plot is a graphical representation of the five-number summary of the data: the minimum value, the first quartile (Q1), the median (Q2), the third quartile (Q3), and the maximum value. The box plot is a useful tool for comparing the distribution of data across different groups. It is also useful for identifying outliers in the data.

```python
# scatterplot
sns.scatterplot(data=df, x="age", y="fare", hue="pclass")
plt.ylim(0, 300)

# plt.ylim is used to set the y-axis limits to 0 and 300.
# from the image we can see that the data is not normally distributed. This is because the data is skewed to the right.
```

```python
sns.pairplot(df, hue="survived")
```

```python
# stacked barchart/plot
df.groupby(["pclass", "sex"])["survived"].mean().unstack().plot(
    kind="bar", stacked=True
)
```

```python
df.groupby(["pclass", "who"])["survived"].mean().unstack().plot(
    kind="bar", stacked=True
)
# for changing position of the xticks in the bar plot we use the following code
# plt.xticks(rotation=0)
```

```python
# swarmplot
sns.swarmplot(data=df, x="survived", y="fare", size=0.7)
```

```python
sns.swarmplot(data=df, x="survived", y="age", hue="who", size=3)
# swarmplot is better in compariosn to scatterplot because it is more informative and it shows the distribution of the data.
# It is also more visually appealing.
```

```python
# another plot which can be used instead of swarm plot is striplot
sns.stripplot(data=df, x="survived", y="fare", hue="who")
# this will plot a strip plot which is similar to swarm plot but with a line connecting each point to the median of the data
```

```python
# catplot
sns.catplot(data=df, kind="point", x="survived", y="age", hue="embarked")
```

```python
# LM plot (to check the relation of two numeric variables)
sns.lmplot(data=df, x="age", y="fare")
```

```python
sns.lmplot(data=df, x="age", y="fare", hue="pclass")
plt.ylim(0, 300)
plt.xlim(0, 50)
# ylim is used to set the y-axis limits
# we can also use xlim to set the x-axis limits

# lmplot is a function from the seaborn library that creates a regression plot which means it will plot the best fit line for the data i.e, the line that best fits the data points.
# The x and y parameters are used to specify the column names of the data that we want to plot. The data parameter is used to specify the data that we want to plot.
# The plt.ylim and plt.xlim functions are used to set the y-axis and x-axis limits respectively. The plt.show() function is used to display the plot.
```

```python
# another way to perform the lm plot with set limits
sns.lmplot(data=df[df["fare"] < 300], x="age", y="fare", hue="pclass")
```

```python
# how to add a column and make a age group
df["age_group"] = pd.cut(
    df["age"],
    [0, 10, 18, 30, 50, 80],
    labels=["child", "adolescent", "young adult", "adult", "senior"],
    include_lowest=True,
)
```

```python
df["family_size"] = df["parch"] + df["sibsp"]


bins = [0, 2, 4, 6, 8]  # Define bins as per your needs
labels = ["Small", "Medium", "Large", "Very Large"]

df["family_size_category"] = pd.cut(
    df["family_size"], bins=bins, labels=labels, right=False
)
```

```python
df.head(10)
```

```python
colors = ["red", "green", "blue", "orange"]  # Adjust colors as needed
df.groupby("family_size_category")["family_size"].mean().plot(kind="bar", color=colors)
```
