# Simple Linear Regression

1. Relationship between two variables is modeled using a straight line, allowing us to predict the value of one variable based on the other.
2. It assumes that there is a linear relationship between the independent and dependent variables, which can be represented mathematically.
3. It is important to note that the model minimizes the sum of the squared differences between the observed values and the values predicted by the linear equation.

y= a + bx

a = constant/intercept (It is also often called as Beta 0)\
b = function / slope of the x (It is also often called as Beta 1)\
y is the predicted value based on the independent variable x.\
x = independent variable\
y = dependent variable

# Multiple Linear Regression

1. In multiple linear regression, the relationship between two or more independent variables and a dependent variable is modeled using a linear equation, allowing for predictions based on multiple inputs.
2. It assumes that there is a linear relationship between the independent variables and the dependent variable, which can be represented mathematically.
3. It is important to note that the model minimizes the sum of the squared differences between the observed values and the values predicted by the linear equation.
4. In multiple linear regression, the coefficients of the independent variables represent the change in the dependent variable for each unit change in the respective independent variable, holding all other variables constant.

y = β0 + β1x1 + β2x2 + ... + βnxn

where,
1. y is the predicted value,
2. β0 is the intercept, β1, β2, ..., βn are the coefficients for each independent variable
3. x1, x2, ..., xn,
4. n is the number of independent variables.
   
In multiple linear regression, it is crucial to assess the model's performance using metrics such as R-squared, adjusted R-squared, and root mean squared error (RMSE) to ensure that the model effectively captures the relationship between the variables.

# Decision Tree Classification

It is a supervised learning algorithm used for classification tasks. It works by splitting the dataset into subsets based on feature values, creating a tree-like model of decisions. Each node in the tree represents a feature (or attribute), and each branch represents a decision rule, leading to the final classification at the leaf nodes.This classification model is based on probability estimates and is less prone to overfitting compared to other models.\
Example: In a dataset of patients, a decision tree might split on the feature "Age" to determine if a patient is at risk of a certain disease. If "Age" is greater than 50, one branch may lead to a higher risk classification, while another branch for "Age" less than or equal to 50 may lead to a lower risk classification.

# Decision Tree Algorithm

Decision tree are non parametric models that can handle both numerical and categorical data. They are also known as decision trees or decision path trees.

If-then-else-decision rules

Pros:
- Easy to interpret and visualize, making them user-friendly for understanding model decisions. 
- Can handle both numerical and categorical data without needing extensive preprocessing.
- Can capture complex interactions between features through hierarchical splits.
- Robust to outliers, as they focus on the most significant splits in the data.
- Can be used for both classification and regression tasks.
- Can be easily combined with other algorithms to improve performance, such as ensemble methods like Random Forests.
- Require less data cleaning and preparation compared to other algorithms, making them quicker to implement.
- Can provide insight into feature importance, helping to identify which features are most influential in making predictions.
- Can handle missing values effectively, allowing for more flexibility in the dataset.
- Can be adapted to various domains and applications, such as finance, healthcare, and marketing.

Cons:
- Prone to overfitting, especially with complex trees that capture noise in the training data.
- Sensitive to small changes in the data, which can lead to different splits and tree structures.
- May require pruning techniques to avoid overfitting and improve generalization.
- Performance can degrade with an imbalanced dataset, as the model may favor the majority class, leading to biased predictions.
- Interpretability can be challenging for large trees, as the complexity increases with depth and the number of features, making it harder to understand the model's decisions.
- Can be computationally expensive to train large trees, especially with high-dimensional data.

# K-Nearest Neighbors (KNN) Classification

It is a supervised learning algorithm used for classification and regression tasks that identifies the 'k' nearest data points in the feature space to make predictions. It works by calculating the distance between points, typically using Euclidean distance, and assigning the most common class label among the nearest neighbors for classification, or averaging the values for regression.
### Definition of Minkowski Distance
The Minkowski distance is a generalization of both Euclidean and Manhattan distances, defined as the p-th root of the sum of the absolute differences raised to the power of p. It is expressed mathematically as:

$$d(x,y) = \left(\sum_{i=1}^{n} |x_i - y_i|^p\right)^{1/p}$$

where $x$ and $y$ are two vectors of equal dimension, and $p$ is the order of the distance metric. The Minkowski distance is a metric that can be used to measure the distance between two points in a vector space, and it is commonly used in machine learning algorithms such as KNN.


### Pros.
- Simple to understand and implement, making it a good choice for beginners in machine learning.
- Highly effective for classification tasks, especially when dealing with high-dimensional data
- Effective for multi-class classification problems, as it can handle multiple classes without requiring complex modifications.
- Can be used with various distance metrics, allowing flexibility depending on the nature of the data.
- Performs well with small to medium-sized datasets, as the computation of distances is straightforward and efficient for smaller datasets.
- Sensitive to the choice of 'k' and the distance metric, which can significantly impact the performance of the algorithm.


### Cons.
- Can suffer from overfitting if the 'k' value is too high, leading to poor generalization to new data.
- May be less effective for regression tasks, as it does not provide a continuous output.
- Can be sensitive to the choice of distance metric, which can impact the performance of the algorithm.
- Requires careful normalization of data, as features with larger ranges can disproportionately influence distance calculations.
- Sensitive to noise and outliers, which can affect the classification results and lead to inaccurate predictions.May struggle with large datasets, as the computation of distances can become computationally expensive and slow with increasing data size.
