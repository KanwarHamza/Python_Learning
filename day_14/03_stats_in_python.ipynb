# Feature Engineering

The process of extracting useful information from raw data is called feature engineering. It is a crucial step in the data preprocessing pipeline. It starts from planning stage as to what data is needed to be collected and how it will be transformed into meaningful features for modeling.

## All analysis are dependent on three main functions

1. Scale of measurement
2. Data type/Variable type
3. Types of studies 

## 1. Scale of measurement

1. Nominal scale: categorical variables that represent different categories or groups of people or things. They do not have any inherent order or ranking.\ 
   Examples include gender, race, or hair color.
2. Ordinal scale: variables that have a meaningful order or ranking, but the differences between the ranks are not uniform.\ 
   Examples include ranks in a competition, educational levels, or satisfaction ratings.
3. Interval scale: variables that have a meaningful order or ranking, and the differences between the ranks are uniform.
   Examples include temperature, time, or IQ scores.
4. Ratio scale: variables that have a meaningful order, and the ratios between values are meaningful.The zero point on the scale is meaningful and has a true value called absolute zero.\
   Examples include height, weight, or age.

## Variables

### 1. Categorical Variables: 
    These are variables that represent categories or groups, such as Gender, Country, or Color.
Variables that represent categories with a finite number of distinct values, such as Color, Brand, etc.
   
   1.1 Nominal Variable: Variables that represent categories without any intrinsic ordering i.e., e.g., Gender, Country, etc.

   1.2 Ordinal Variable: Variables that represent categories with a meaningful order or ranking, such as Satisfaction Ratings, Education Level, etc.


### 2. Numerical Variables: 
    These are Variables that represent numerical values, such as Age, Height, Weight, etc.

   2.1 Continuous: Variables that can take any value within a given range, such as Temperature, Time, or Distance.

   2.2 Discrete: Variables that can only take specific, distinct values, such as Number of Children, Count of Items, etc.

### Types of Variables
1. Dependent Variables: variables whose values depend on other variables and can be controlled or manipulated in an experiment. 
2. Independent Variables: variables that are manipulated or controlled in an experiment to observe their effect on the dependent variable.

### Time series data vs cross-sectional data

1. Time series data: data collected at different points in time for the same entity or population. 
2. Cross-sectional data: data collected at different points in time for different entities or populations.
