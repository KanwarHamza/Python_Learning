## Hypothesis Testing
Hypothesis testing is a statistical method that uses sample data to evaluate a hypothesis about a population parameter.
It involves formulating a null hypothesis and an alternative hypothesis, then determining the likelihood of observing the sample data under the null hypothesis.
1. Null Hypothesis (H0): The population parameter is equal to a specific value. For example The average height of students in the high school is equal to 160 cm. For example, if the null hypothesis states that the average height of students is 160 cm, we would test this against the alternative hypothesis that suggests a different average height.
2. Alternative Hypothesis (Ha): The population parameter is not equal to a specific value. For example The average height of students in the high school is less than 160 cm. The average height of students in the high school is greater than 160 cm.
## Steps in Hypothesis Testing
1. Formulate the null and alternative hypotheses.
2. Choose a significance level (α).
3. Collect a sample from the population.
4. Calculate the test statistic.
5. Determine the critical region or p-value.
6. Make a decision based on the test statistic and the chosen significance level.
7. Interpret the results and draw conclusions about the population parameter.
# Example: Testing the Mean of a Population
Suppose you want to test whether the average height of students in a high school is greater than 160 cm.
1. Null Hypothesis (H0): The average height of students in the high school is 160 cm or less.
2. Alternative Hypothesis (Ha): The average height of students in the high school is greater than 160 cm.
3. Significance Level (α): 0.05
4. Sample Data: Heights of 30 randomly selected students in the high school (in cm).
5. Test Statistic: t-statistic (t = (sample mean - population mean) / (sample standard deviation / sqrt(sample size)))
6. Critical Region: If the calculated t-statistic falls in the critical region (t > t-critical), reject the null hypothesis.
7. Decision: If the calculated t-statistic falls in the critical region, conclude that there is evidence to support the alternative hypothesis.
8. Conclusion: The average height of students in the high school is greater than 160 cm.
In this example, we used a one-sample t-test to evaluate the hypothesis regarding the average height of students. The results indicate that we have sufficient evidence to suggest that the average height exceeds 160 cm, based on the sample data collected.

# Confidence Interval

Confidence interval is a range of values below and above the sample mean that is likely to contain the true population mean.

CI = Point Estimate ± Critical Value * Standard Error\
CI = 95% means that 95% of the time, the true population mean will fall within the CI.\
CI = 99% means that 99% of the time, the true population mean will fall within the CI.

* x = Sample Mean
* n = Sample Size
* x̄ = Population Mean
* s = Sample Standard Deviation
* σ = Population Standard Deviation
* z = Z-Score
* t = T-Score
* α = Significance Level
* CI = Confidence Interval

CI = x̄ ± z * (s / sqrt(n))

# Step 2 - Homogeneity of variance testing
Homogenity of variance testing is used to check if the variances of two or more groups are equal. This is important because the assumption of homogeneity of variance is one of the assumptions of many statistical tests, such as ANOVA and t-tests. If the variances are not equal, the results of these tests may be biased and unreliable.

1. Levene's test: A test for homogeneity of variance that is robust to outliers. It's a non-parametric test that compares the variances of two or more groups. If the P value is less than 0.05, we reject the null hypothesis that the variances are equal.
2. Bartlett's test: A test for homogeneity of variance that is less robust to outliers than Levene's test. It's a non-parametric test that compares the variances of two or more groups. If the P value is less than 0.05, we reject the null hypothesis that the variances are equal.
3. Brown-Forsythe test: A test for homogeneity of variance that is less robust to outliers than Levene's test. It's a non-parametric test that compares the variances of two or more groups. If the P value is less than 0.05, we reject the null hypothesis that the variances are equal.
4. Fligner-Killeen test: A test for homogeneity of variance that is less robust to outliers than Levene's test. It's a non-parametric test that compares the variances of two or more groups. If the P value is less than 0.05, we reject the null hypothesis that the variances are equal.
5. Welch's test: A test for homogeneity of variance that is less robust to outliers than Levene's test. It's a parametric test that compares the variances of two or more groups. If the P value is less than 0.05, we reject the null hypothesis that the variances are equal.

```python
# now lets do chi-sq test for homogenityor independece
import numpy as np
import scipy.stats as stats
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import chi2_contingency
from itertools import combinations

# we will use 'chi2_contingency' function from scipy.stats module to perform chi-sq test for homogenity.
# the variables will include 'sex', 'embarked', 'pclass', 'survivved'


def pairwise_chi_square_homogeneity_tests(data, variables, alpha=0.05):
    pairs = list(combinations(variables, 2))
    for var1, var2 in pairs:
        contingency_table = pd.crosstab(data[var1], data[var2])
        chi2, p, dof, expected = chi2_contingency(contingency_table)
        p_label = f"< {alpha}" if p < alpha else f">= {alpha}"
        significant = p < alpha
        print(f"Chi-square test for homogeneity between '{var1}' and '{var2}':")
        print(f"  Chi-square statistic: {chi2:.4f}")
        print(f"  p-value: {p_label}")
        print(f"  Degrees of freedom: {dof}")
        print(f"  Significant: {significant}")
        print("  Contingency Table:")
        print(contingency_table)
        print("-" * 50)


# Example usage:
pairwise_chi_square_homogeneity_tests(df, ["sex", "embarked", "pclass", "survived"])


# for grouped bar plot we will use seaborn library
import seaborn as sns
import matplotlib.pyplot as plt


def plot_grouped_bar(data, var1, var2):
    sns.countplot(data=data, x=var1, hue=var2)
    plt.title(f"Distribution of {var1} grouped by {var2}")
    plt.show()


# Example usage:
plot_grouped_bar(df, "sex", "survived")
plot_grouped_bar(df, "embarked", "survived")
plot_grouped_bar(df, "pclass", "sex")
plot_grouped_bar(df, "pclass", "survived")


# for mosaic plot we will use seaborn library
from statsmodels.graphics.mosaicplot import mosaic
import matplotlib.pyplot as plt


def plot_mosaic(data, var1, var2):
    mosaic(data, [var1, var2])
    plt.title(f"Mosaic Plot of {var1} and {var2}")
    plt.show()


# Example usage:
plot_mosaic(df, "sex", "survived")
plot_mosaic(df, "embarked", "survived")
plot_mosaic(df, "pclass", "sex")
plot_mosaic(df, "pclass", "survived")
```

```python
# Anova test
# for anova using categories of sex, survived and embarked and the continuous variables will be fare and age
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats

import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm


import statsmodels.api as sm

from statsmodels.stats.multicomp import pairwise_tukeyhsd

df.head()


def get_simplified_anova_results(model_formula, data, alpha=0.05):
    model = ols(model_formula, data=data).fit()
    anova_results = anova_lm(model)
    # Rename columns for clarity
    anova_results = anova_results.rename(
        columns={
            "sum_sq": "Sum of Squares",
            "df": "Degrees of Freedom",
            "F": "F-Statistic",
            "PR(>F)": "p-value",
        }
    )
    # Replace p-values with categorical labels
    anova_results["p-value"] = anova_results["p-value"].apply(
        lambda p: f"< {alpha}" if p < alpha else f">= {alpha}"
    )
    # Add a column for significance as boolean
    anova_results["Significant"] = anova_results["p-value"].apply(
        lambda x: True if x.startswith("<") else False
    )
    return anova_results


# Example usage:
anova_fare = get_simplified_anova_results(
    "fare ~ C(sex) + C(survived) + C(embarked)", df
)
print("Simplified ANOVA results for fare:")
print(anova_fare)

anova_age = get_simplified_anova_results("age ~ C(sex) + C(survived) + C(embarked)", df)
print("\nSimplified ANOVA results for age:")
print(anova_age)

# to make a visual representation of the findings

import pandas as pd
import matplotlib.pyplot as plt


def plot_anova_results(anova_data, title):
    df = pd.DataFrame(anova_data)
    df["F-Statistic"] = df["F-Statistic"].fillna(0)

    colors = ["green" if sig else "gray" for sig in df["Significant"]]

    plt.figure(figsize=(8, 5))
    bars = plt.bar(df["Factor"], df["F-Statistic"], color=colors)

    plt.title(title)
    plt.ylabel("F-Statistic")
    plt.xlabel("Factor")

    for bar, sig in zip(bars, df["Significant"]):
        height = bar.get_height()
        if sig:
            plt.text(
                bar.get_x() + bar.get_width() / 2,
                height + 1,
                "*",
                ha="center",
                va="bottom",
                fontsize=14,
                color="red",
            )

    plt.ylim(0, df["F-Statistic"].max() * 1.2)
    plt.show()


# Example data
anova_fare_data = {
    "Factor": ["C(sex)", "C(survived)", "C(embarked)", "Residual"],
    "F-Statistic": [79.180114, 53.071584, 32.237228, None],
    "Significant": [True, True, True, False],
}

anova_age_data = {
    "Factor": ["C(sex)", "C(survived)", "C(embarked)", "Residual"],
    "F-Statistic": [2.839308, 0.031494, 0.398985, None],
    "Significant": [False, False, False, False],
}

# Plot for fare
plot_anova_results(anova_fare_data, "ANOVA Results for Fare")

# Plot for age
plot_anova_results(anova_age_data, "ANOVA Results for Age")
```

```python
# second lets calcuate correlation between two variables
# between age and fare
# Since it is non parametric, we can't use pearson correlation
# lets use spearman correlation

import numpy as np
from scipy.stats import spearmanr

# Calculate Spearman correlation matrix and p-value matrix
corr_matrix, p_matrix = spearmanr(df[["age", "fare", "survived"]])

# Variable names
variables = ["age", "fare", "survived"]

# Print correlation matrix with labels
print("Spearman Correlation Matrix:")
print(pd.DataFrame(corr_matrix, index=variables, columns=variables).round(3))
print("\nP-value Matrix:")
print(
    pd.DataFrame(p_matrix, index=variables, columns=variables).map(lambda x: f"{x:.2f}")
)

# Interpret significance for each pair
alpha = 0.05
for i in range(len(variables)):
    for j in range(i + 1, len(variables)):
        var1, var2 = variables[i], variables[j]
        corr = corr_matrix[i, j]
        p_val = p_matrix[i, j]
        significance = "significant" if p_val < alpha else "not significant"
        print(
            f"The correlation between {var1} and {var2} is {corr:.3f} and is {significance} (p = {p_val:.2f})."
        )
```
