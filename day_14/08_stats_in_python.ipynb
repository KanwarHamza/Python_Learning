# Critical Region
It is also called as region of rejection in statistical hypothesis testing. In this region, we decide whether to reject the null hypothesis based on the observed data. Typically, the critical region corresponds to the extreme values of the test statistic, which lie beyond a certain threshold determined by the significance level.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, t


# 1. Code to show Critical Region for a one-tailed t-test
def plot_critical_region(alpha=0.05, df=29):
    x = np.linspace(-4, 4, 1000)
    y = t.pdf(x, df)
    t_critical = t.ppf(1 - alpha, df)

    plt.plot(x, y, label="t-distribution df=29")
    plt.fill_between(
        x,
        0,
        y,
        where=(x >= t_critical),
        color="red",
        alpha=0.5,
        label="Critical Region (alpha=0.05)",
    )
    plt.axvline(
        t_critical,
        color="red",
        linestyle="--",
        label=f"Critical value = {t_critical:.2f}",
    )
    plt.title("Critical Region for One-tailed t-test")
    plt.xlabel("t value")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.show()
```

# Significant Level
The sig level is a measure of the importance or impact of a particular statistic or event. It is commonly used in hypothesis testing to determine whether to reject the null hypothesis. It is denoted by a significance level, often represented as alpha (Î±). Common choices include 0.05, 0.01, and 0.10.

```python
# 2. Code to show Significance Level on a normal distribution curve
def plot_significance_level(alpha=0.05):
    x = np.linspace(-4, 4, 1000)
    y = norm.pdf(x)
    z_critical = norm.ppf(1 - alpha)

    plt.plot(x, y, label="Standard Normal Distribution")
    plt.fill_between(
        x,
        0,
        y,
        where=(x >= z_critical),
        color="orange",
        alpha=0.5,
        label=f"Significance Level (alpha={alpha})",
    )
    plt.axvline(
        z_critical,
        color="orange",
        linestyle="--",
        label=f"Critical z = {z_critical:.2f}",
    )
    plt.title("Significance Level on Normal Distribution")
    plt.xlabel("z value")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.show()
```

# P-value
It is a measure of the strength of evidence against the null hypothesis in statistical hypothesis testing. The smaller the p-value, the stronger the evidence against the null hypothesis. Conversely, a larger p-value suggests that the observed data is consistent with the null hypothesis, indicating insufficient evidence to reject it. In general, a p-value threshold (commonly set at 0.05) is used to determine statistical significance, where a p-value below this threshold indicates that the null hypothesis can be rejected.

```python
# 3. Code to show p-value for a given test statistic on normal distribution
def plot_p_value(test_stat=1.5):
    x = np.linspace(-4, 4, 1000)
    y = norm.pdf(x)
    p_value = 1 - norm.cdf(test_stat)

    plt.plot(x, y, label="Standard Normal Distribution")
    plt.fill_between(
        x,
        0,
        y,
        where=(x >= test_stat),
        color="green",
        alpha=0.5,
        label=f"p-value = {p_value:.4f}",
    )
    plt.axvline(
        test_stat, color="green", linestyle="--", label=f"Test Statistic = {test_stat}"
    )
    plt.title("p-value Visualization on Normal Distribution")
    plt.xlabel("z value")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.show()


if __name__ == "__main__":
    plot_critical_region()
    plot_significance_level()
    plot_p_value()
```

# Type-I and Type-II Errors and its removal
Type-I errors, also known as false positives, occur when a null hypothesis is incorrectly rejected, suggesting an effect or difference exists when it actually does not. 
Type-II errors, or false negatives, happen when a null hypothesis is not rejected when it should be, indicating no effect or difference exists when it actually does. 
To minimize these errors, researchers can increase sample sizes, use more powerful statistical tests, and ensure proper experimental design and data collection methods. Additionally, conducting pilot studies can help identify potential issues before the main experiment, allowing for adjustments to be made to reduce the likelihood of both Type-I and Type-II errors.

In statical analysis, Type-I and Type-II errors play a crucial role in hypothesis testing and decision-making. To reduce these errors, researchers can implement the following strategies:
1. Utilize a significance level (alpha) that balances the risk of Type-I errors while considering the context of the study.
2. Employ power analysis to determine the appropriate sample size needed to achieve a desired level of statistical power, thereby reducing the risk of Type-II errors.
3. Employ multiple testing correction techniques to control the overall rate of Type-I errors, especially when conducting multiple hypothesis tests simultaneously.
4. Implement robust statistical methods that account for potential confounding variables and ensure valid comparisons between groups or conditions.
5. Conduct thorough data cleaning and preprocessing to ensure the quality and reliability of the data, reducing the likelihood of Type-II errors due to missing or erroneous data.
6. Use appropriate statistical tests that are sensitive to the nature of the data and the research question, minimizing the risk of Type-I errors.
7. Conduct sensitivity analyses to assess the robustness of the results to different assumptions or alternative hypotheses, thereby reducing the risk of Type-II errors.
8. Collaborate with domain experts and stakeholders to ensure that the research question and hypotheses are well-defined and relevant to the study's objectives.
9. Continuously monitor and evaluate the study's performance, adjusting the statistical analysis and experimental design as needed to minimize the risk of Type-I and Type-II errors over time.
By implementing these strategies, researchers can effectively reduce the likelihood of Type-I and Type-II errors, leading to more reliable and valid conclusions in their research.

# Random Variables

A random variable is a variable whose possible values are outcomes of a random phenomenon. It can take on different values based on the result of a random event, such as a coin toss or the roll of a die. Random variables are often classified into two categories: discrete and continuous. Discrete random variables can take on a countable number of values, while continuous random variables can take on an infinite number of values within a given range.

It is based of two types of distributions:
1. Discrete Distribution
2. Probability Distribution

# Discrete Distribution
Discrete distribution describes the probability distribution of a discrete random variable, which can take on specific, distinct values. Examples include the binomial distribution, Poisson distribution, and geometric distribution. These distributions provide a way to model scenarios where outcomes are finite and countable, such as the number of successes in a fixed number of trials or the number of arrivals at a bus stop in a given time frame.

# Probability Distribution
Probability distribution describes the likelihood of different outcomes for a continuous random variable. Examples include the normal distribution, exponential distribution, and uniform distribution. These distributions are useful for modeling scenarios where outcomes can take on any value within a specified range, such as the heights of individuals in a population or the time it takes for a computer to complete a task of a certain task.

### Types of Probability Distributions
1. Discrete Probability Distribution
2. Continuous Probability Distribution

Discrete Probability Distribution describes the probability distribution of a discrete random variable. It assigns probabilities to each of the possible values that the variable can take. Examples of discrete probability distributions include the binomial distribution, which models the number of successes in a fixed number of independent Bernoulli trials, and the Poisson distribution, which models the number of events occurring in a fixed interval of time or space.

Continuous Probability Distribution describes the probability distribution of a continuous random variable. It assigns probabilities to ranges of values rather than specific values. Examples of continuous probability distributions include the normal distribution, which is characterized by its bell-shaped curve and is used to model phenomena such as heights or test scores, and the exponential distribution, which is often used to model the time until an event occurs, such as the time until a radioactive particle decays.

### Discrete Probability Distribution
It can be further divided into various types, including the binomial, Poisson, and geometric distributions.

#### Binomial Distribution
The binomial distribution is a discrete probability distribution that models the number of successes in a fixed number of independent Bernoulli trials. It is characterized by two parameters: the number of trials (n) and the probability of success (p) in each trial. The binomial distribution is often used to model scenarios where there are only two possible outcomes, such as the number of heads in a series of coin flips or the number of defective items in a batch of products.

#### Poisson Distribution
The Poisson distribution is a discrete probability distribution that models the number of events occurring in a fixed interval of time or space. It is often used in scenarios where events occur independently and the average rate of occurrence is constant over time. Examples include modeling the number of phone calls received at a call center in an hour or the number of emails received in a day.

#### Geometric Distribution
The geometric distribution is a discrete probability distribution that models the number of trials needed to get the first success in a sequence of independent Bernoulli trials. It is often used to model scenarios where there is a fixed probability of success on each trial, such as the number of coin flips needed to get heads or the number of attempts needed to find a needle in a haystack.

### Continuous Probability Distribution
It can be further divided into various types, including the normal, exponential, and uniform distributions.

#### Normal Distribution
The normal distribution, also known as the Gaussian distribution, is a continuous probability distribution that is widely used in statistics and probability theory. It is characterized by its bell-shaped curve and is often used to model real-valued random variables whose distributions are not known.
